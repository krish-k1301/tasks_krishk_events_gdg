So, make a new column called "Lost Weight". Use the following formula to populate every row of it:



```
Weight - (Shucked + Viscera + Shell)
```


If this operation gives a value of Lost Weight that is less than or equal to zero, then assign zero to that row's Lost Weight value. Otherwise, assign one to it.

Also, ensure that this column is added in between Shell Weight and Age.

````python
# write code here
df.loc[:, 'Lost weight'] = df['Weight'] - (
    df['Shucked Weight'] + df['Viscera Weight'] + df['Shell Weight']
)
````

## One Hot Encoding vs Label Encoding: A Short Report

When working with machine learning models, categorical data must be converted into numerical form. Two popular methods for this are Label Encoding and One Hot Encoding. Both techniques serve the same purpose, but they work in different ways and are used in different situations.

### Label Encoding

Label Encoding assigns an integer value to each unique category in a column. For example, if a column named Color has values Red, Blue and Green, a label encoder might convert them into numeric form like this:

Red → 0
Blue → 1
Green → 2

This method is simple and efficient. It works well when the categories have a natural order. Examples include:

Small < Medium < Large
Cold < Warm < Hot

In such cases, the numbers represent meaningful relationships between categories. However, Label Encoding becomes a problem when the categories have no order. For example, if we encode Male as 0 and Female as 1, the model might interpret 1 as greater than 0, even though gender does not have a numeric relationship. This can mislead models like linear regression or SVMs.

### One Hot Encoding

One Hot Encoding solves this issue by creating separate binary columns for each category. Using the same Color example, the values would be converted into three columns:

Color_Red:    1 or 0
Color_Blue:   1 or 0
Color_Green:  1 or 0

If the color is Blue, the encoded row becomes:

Color_Red = 0
Color_Blue = 1
Color_Green = 0

Since each category gets its own column, there is no implied order. One Hot Encoding is therefore preferred when the categories are nominal and have no ranking or natural hierarchy.

### When To Use Which

Label Encoding is useful when the categorical feature has an order and the model should understand that order. Examples include education level, size categories or temperature ranges.
One Hot Encoding is used when the categories are independent and do not follow any kind of ordering. Examples include gender, color, country or product type.

### Drawbacks

Label Encoding can introduce unintended relationships between categories when there is no natural order.
One Hot Encoding increases the number of columns in the dataset, especially when the categorical feature has many unique values. This can make the dataset harder to work with and can slow down some models.

### Summary

Label Encoding assigns a numeric label to each category and works best with ordered categories. One Hot Encoding converts each category into a separate binary column and works best with unordered categories. Choosing between the two depends on the type of categorical data and the machine learning model being used.
